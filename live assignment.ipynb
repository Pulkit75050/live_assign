```python
""" Q3. Explain the following terms
 The difference in np.array, np.asarray and np.asanyarray
 The difference between Deep copy and shallow copy"""

"""
SOL.

np.array: This function creates a new NumPy array from an input object (like a list, tuple, or another array).
np.asarray: Similar to np.array, this function converts an input object to a NumPy array.
np.asanyarray: This function is similar to np.asarray in that it avoids copying the input if it’s already a NumPy array.

A shallow copy creates a new object but does not create copies of the nested objects within the original object. Instead, it just copies the references to these nested objects. 
So, if you modify a nested object in the original,the change will be reflected in the shallow copy, and vice versa.

A deep copy creates a new object and also creates separate copies of all nested objects found within the original. 
Changes to any nested objects in the original will not affect the copied objects in the deep copy, and vice versa.
"""
```

```python
""" Q5. Create a NumPy array with random integers between 1 and 10 of shape (5, 6). After creating the array perform the following operations:
 a)Extract all even integers from array.
 b)Extract all odd integers from array."""

# Create a 5x6 array with random integers between 1 and 10
random_array = np.random.randint(1, 11, size=(5, 6))

# Extract even integers from the array
even_integers = random_array[random_array % 2 == 0]

# Extract odd integers from the array
odd_integers = random_array[random_array % 2 != 0]

print(random_array)
print(even_integers)
print(odd_integers)
```

```python
""" Q6. Create a 3D NumPy array of shape (3, 3, 3) containing random integers between 1 and 10. Perform the following operations:
 a) Find the indices of the maximum values along each depth level (third axis).
 b) Perform element-wise multiplication of between both array."""

# Create a 3D array of shape (3, 3, 3) with random integers between 1 and 10
random_3d_array = np.random.randint(1, 11, size=(3, 3, 3))

# Find indices of maximum values along the third axis
max_indices = np.unravel_index(random_3d_array.argmax(axis=2), random_3d_array.shape)

# Perform element-wise multiplication
elementwise_product = random_3d_array * random_3d_array

print(random_3d_array)
print(max_indices)
print(elementwise_product)
```

```python
""" Q14.  Create a time-series dataset in a Pandas DataFrame with columns: 'Date', 'Temperature', 'Humidity' and Perform the following tasks using Matplotlib:
  a) Plot the 'Temperature' and 'Humidity' on the same plot with different y-axes (left y-axis for 'Temperature' and
  right y-axis for 'Humidity').
 b) Label the x-axis as 'Date'.
c) Set the title of the plot as 'Temperature and Humidity Over Time'."""

# Create a time-series dataset
date_range = pd.date_range(start='2024-01-01', periods=100, freq='D')
temperature = np.random.uniform(low=20, high=30, size=(100,))
humidity = np.random.uniform(low=40, high=60, size=(100,))
df = pd.DataFrame({'Date': date_range, 'Temperature': temperature, 'Humidity': humidity})

# Plotting
fig, ax1 = plt.subplots()

# Plot 'Temperature' on the left y-axis
color = 'tab:red'
ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature', color=color)
ax1.plot(df['Date'], df['Temperature'], color=color)
ax1.tick_params(axis='y', labelcolor=color)

# Create a second y-axis for 'Humidity'
ax2 = ax1.twinx()
color = 'tab:blue'
ax2.set_ylabel('Humidity', color=color)
ax2.plot(df['Date'], df['Humidity'], color=color)
ax2.tick_params(axis='y', labelcolor=color)

# Formatting Date on x-axis
date_form = DateFormatter("%Y-%m-%d")
ax1.xaxis.set_major_formatter(date_form)

# Set the title
plt.title('Temperature and Humidity Over Time')

# Show plot with grid
plt.grid(True)
plt.show()
```

```python
""" Q15.  Create a NumPy array data containing 1000 samples from a normal distribution. Perform the following tasks using Matplotlib:
 a) Plot a histogram of the data with 30 bins.
 b) Overlay a line plot representing the normal distribution's probability density function (PDF).
 c) Label the x-axis as 'Value' and the y-axis as 'Frequency/Probability'.
 d) Set the title of the plot as 'Histogram with PDF Overlay'."""

# Generate 1000 samples from a normal distribution
mean = 0  # Mean (center of the distribution)
std_dev = 1  # Standard deviation (spread of the distribution)
data = np.random.normal(mean, std_dev, size=1000)

# Plot a histogram with 30 bins
plt.hist(data, bins=30, density=True, alpha=0.6, color='blue', edgecolor='black', label='Histogram')

# Overlay the PDF (probability density function)
x_values = np.linspace(min(data), max(data), 1000)
pdf_values = (1 / (std_dev * np.sqrt(2 * np.pi))) * np.exp(-(x_values - mean)**2 / (2 * std_dev**2))
plt.plot(x_values, pdf_values, color='red', linewidth=2, label='PDF')

# Label axes and set title
plt.xlabel('Value')
plt.ylabel('Frequency/Probability')

# Display legend
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```

```python
""" Q17.  Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the 
origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'."""

# Generate two random arrays
x = np.random.uniform(-1, 1, 100)
y = np.random.uniform(-1, 1, 100)

# Determine quadrant for each point
quadrant = np.zeros(100)
quadrant[(x > 0) & (y > 0)] = 1  # Quadrant I
quadrant[(x < 0) & (y > 0)] = 2  # Quadrant II
quadrant[(x < 0) & (y < 0)] = 3  # Quadrant III
quadrant[(x > 0) & (y < 0)] = 4  # Quadrant IV

# Create a DataFrame for plotting
data = pd.DataFrame({'x': x, 'y': y, 'Quadrant': quadrant})

# Set up the scatter plot
sns.set(style='whitegrid')
plt.figure(figsize=(8, 6))
sns.scatterplot(x='x', y='y', hue='Quadrant', data=data, palette='Set1', legend='full')

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Quadrant-wise Scatter Plot')

# Show the legend
plt.legend(title='Quadrant', loc='upper right')

# Show the plot
plt.grid(True)
plt.show()
```

```python
""" Q19.  Using Bokeh, generate a bar chart of randomly generated categorical data, color bars based on their values, add hover tooltips to display exact values, 
label the axes, and set the title as 'Random Categorical Bar Chart'."""

# Prepare the output for a Jupyter notebook (comment out if not using a notebook)
output_notebook()

# Generate random categorical data
categories = ['A', 'B', 'C', 'D', 'E']
values = np.random.randint(1, 20, size=len(categories))

# Create a ColumnDataSource
source = ColumnDataSource(data={'categories': categories, 'values': values})

# Set up the plot
p = figure(x_range=categories, title='Random Categorical Bar Chart')
p.vbar(x='categories', top='values', source=source, width=0.6,
       line_color='white', fill_color=factor_cmap('categories', palette=Category20[5], factors=categories))

# Add hover tooltips
hover = HoverTool()
hover.tooltips = [('Category', '@categories'), ('Value', '@values')]
p.add_tools(hover)

# Label axes
p.xaxis.axis_label = 'Categories'
p.yaxis.axis_label = 'Values'

# Show the plot
show(p)
```
